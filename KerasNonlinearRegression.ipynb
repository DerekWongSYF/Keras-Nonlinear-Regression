```python
from keras.models import Sequential
from keras.layers import Dense
import matplotlib.pyplot as plt
import numpy
from sklearn.metrics import r2_score
%matplotlib inline

seed = 7
numpy.random.seed(seed)
dataset = numpy.genfromtxt("nonlinear.csv", delimiter=",", skip_header=True)
dataset2 = numpy.genfromtxt("validation.csv", delimiter=",", skip_header=True)
# split into input (X) and output (Y) variables
X = dataset[:,0:5]
Y = dataset[:,5]

X2 = dataset2[:,0:5]
Y2 = dataset2[:,5]
"""
print (X)
print ("XEnd")
print (Y)
print ("YEnd")
print (X2)
print ("X2End")
"""

# create model
model = Sequential()
model.add(Dense(50, input_dim=5, init='uniform', activation='tanh'))
#model.add(Dense(40, init='uniform', activation='tanh'))
#model.add(Dense(30, init='uniform', activation='tanh'))
#model.add(Dense(20, init='uniform', activation='tanh'))
#model.add(Dense(10, init='uniform', activation='sigmoid'))
#model.add(Dense(5, init='uniform', activation='sigmoid'))
model.add(Dense(1, init='uniform', activation='linear'))
# Compile model
model.compile(loss='mse', optimizer='adam', metrics=['accuracy'])
# Fit the model
model.fit(X, Y, nb_epoch=5000, batch_size=10,  verbose=2)
# calculate predictions
PredTestSet = model.predict(X)
PredValSet = model.predict(X2)

#print(predictions)

numpy.savetxt("testresults.csv", PredTestSet, delimiter=",")
numpy.savetxt("valresults.csv", PredValSet, delimiter=",")
```

```python
dataset3 = numpy.genfromtxt("testresults.csv", delimiter=",")
plt.plot(Y,dataset3,'ro')
#print(Y2.shape)
#print(dataset3.shape)

Testr2value = r2_score(Y,dataset3)
print(Testr2value)



```

```python
dataset4 = numpy.genfromtxt("valresults.csv", delimiter=",")
plt.plot(Y2,dataset4,'ro')

Valr2value = r2_score(Y2,dataset4)
print(Valr2value)
```
